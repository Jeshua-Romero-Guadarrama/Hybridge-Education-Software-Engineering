### 11. Imprime el número total de valores NA que haya en cada columna. Hazlo primero para la columna `Metascore` y después utiliza un ciclo for para hacerlo para todas las columnas

```python
conteos_na = {}
for columna in df.columns:
    conteos_na[columna] = df[columna].isna().sum()
for columna, conteo in conteos_na.items():
    print(f"{columna}: {conteo} valores NA")
```

### 21. Obten los valores únicos de la variable Rating y la frecuencia total de cada uno de estos valores. Posteriormente crea un nuevo dataframe con esos valores

```python
conteos_rating         = df['rating'].value_counts().reset_index()
conteos_rating.columns = ['Rating', 'Frecuencia']
conteos_rating
```

### 22. Ordena el nuevo dataframe por la variable `rating`

```python
conteos_rating = conteos_rating.sort_values('Rating')
conteos_rating
```

### 23. Crea una gráfica de barras con este nuevo dataframe ordenado

```python
conteos_rating.plot(kind = 'bar', x = 'Rating', y = 'Frecuencia', legend = False, figsize = (12, 4), color = 'skyblue')
plt.title('Frecuencia de Ratings')
plt.xlabel('Rating')
plt.ylabel('Frecuencia')
plt.show()
```

### 24. Crea la matriz de correlación del dataframe de películas

```python
columnas_numericas = df[['year', 'runtime_minutes', 'rating', 'votes', 'revenue_millions', 'metascore']]
matriz_correlacion = columnas_numericas.corr()
matriz_correlacion
```

### 25. Grafica la matriz de correlación utilizando `matshow()`

```python
plt.figure(figsize = (10, 4))
sns.heatmap(matriz_correlacion, annot = True, fmt = ".2f", cmap = 'coolwarm', cbar = True)
plt.title('Matriz de Correlación de las Variables Numéricas')
plt.show()
```
